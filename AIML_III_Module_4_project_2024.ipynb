You need to run gradient descent with three different learning rates (low_gamma, mid_gamma, high_gamma) and experiment with an additional rate (insane_gamma).

First, let's define the gradient_descent function:

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.01, precision=0.001, max_iter=1000):
    new_x = initial_guess
    x_list = [new_x]
    derivative_list = [derivative_func(new_x)]
    for _ in range(max_iter):
        previous_x = new_x
        new_x = previous_x - multiplier * derivative_func(previous_x)
        x_list.append(new_x)
        derivative_list.append(derivative_func(new_x))
        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    return new_x, x_list, derivative_list
```

 Firstly, Define the Symbolic Function and Its Derivatives

```python
import numpy as np
import sympy as sp

# Define symbolic variables
a, b = sp.symbols('a b')

# Define the function f(x, y) = 13 - x^2 - y^2 + 1
f_sym = 13 - a**2 - b**2 + 1

# Compute the partial derivatives
df_da = sp.diff(f_sym, a)
df_db = sp.diff(f_sym, b)

# Define numerical versions of the partial derivatives
def fpx(x, y):
    return float(df_da.evalf(subs={a: x, b: y}))

def fpy(x, y):
    return float(df_db.evalf(subs={a: x, b: y}))
```

Batch Gradient Descent Implementation

```python
# Batch Gradient Descent
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])  # initial guess

for n in range(max_iter):
    gradient_x = fpx(params[0], params[1])
    gradient_y = fpy(params[0], params[1])
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients

# Results
print('Values in gradient array:', gradients)
print('Minimum occurs at x value of:', params[0])
print('Minimum occurs at y value of:', params[1])
print('The cost is:', f_sym.evalf(subs={a: params[0], b: params[1]}))
```

```python
import numpy as np

def mse(y, y_hat):
    # Ensure y and y_hat are numpy arrays
    y = np.array(y)
    y_hat = np.array(y_hat)

    # Compute the Mean Squared Error
    mse_calc = np.mean((y - y_hat)**2)
    return mse_calc
```
