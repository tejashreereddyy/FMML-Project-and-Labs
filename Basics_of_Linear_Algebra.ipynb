MULTI_DOT METHOD

The numpy.linalg.multi_dot method computes the dot product of two or more arrays in an optimized manner.numpy.linalg.multi_dot allows you to multiply multiple matrices together in the most efficient way. It accepts a list of 2D arrays (matrices) and computes their product.

```python
import numpy as np

# Define the matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = np.array([[9, 10], [11, 12]])

# Compute the product using multi_dot
result = np.linalg.multi_dot([A, B, C])

# Print the result
print(result)
```

INNER PRODUCT AND OUTER PRODUCT

Inner Product

The inner product (or dot product) of two vectors results in a scalar value and is calculated by multiplying corresponding elements and summing the results.

Outer Product

The outer product of two vectors results in a matrix, where each element is the product of elements from the two input vectors.

```python
import numpy as np

# Define two vectors
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Compute inner product
inner_product = np.dot(a, b)

# Compute outer product
outer_product = np.outer(a, b)

# Print results
print("Inner Product:", inner_product)
print("Outer Product:\n", outer_product)
```

Matrix Operations

 The several basic matrix operations performed using NumPy, demonstrating some fundamental linear algebra concepts are as follows:

Matrix Addition,
Matrix Subtraction,
Matrix Multiplication,
Element-wise Multiplication,
Transpose of a Matrix,
Inverse of a Matrix,
Determinant of a Matrix and
Eigenvalues and Eigenvectors

```python
import numpy as np

# Define two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# 1. Matrix Addition
addition = A + B

# 2. Matrix Subtraction
subtraction = A - B

# 3. Matrix Multiplication
multiplication = np.dot(A, B)

# 4. Element-wise Multiplication
element_wise_multiplication = A * B

# 5. Transpose of a Matrix
transpose = np.transpose(A)

# 6. Inverse of a Matrix
# Note: The matrix must be square and non-singular
inverse = np.linalg.inv(A)

# 7. Determinant of a Matrix
determinant = np.linalg.det(A)

# 8. Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Print results
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("\nMatrix Addition:\n", addition)
print("Matrix Subtraction:\n", subtraction)
print("Matrix Multiplication:\n", multiplication)
print("Element-wise Multiplication:\n", element_wise_multiplication)
print("Transpose of A:\n", transpose)
print("Inverse of A:\n", inverse)
print("Determinant of A:\n", determinant)
print("Eigenvalues of A:\n", eigenvalues)
print("Eigenvectors of A:\n", eigenvectors)
```

Matrix Power

The matrix power operation raises a square matrix to a given integer power. This operation is performed using repeated matrix multiplication, which is useful in various applications like solving linear differential equations and Markov chains.

```python
import numpy as np

# Define a square matrix
A = np.array([[1, 2], [3, 4]])

# Compute the power of the matrix
# For example, A^3 (A raised to the power of 3)
power = np.linalg.matrix_power(A, 3)

# Print the result
print("Matrix A:\n", A)
print("\nA^3:\n", power)
```

 Solving and Plotting Linear Equations


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the coefficients of the linear equations
A = np.array([[2, 1], [1, -1]])
b = np.array([4, 1])

# Solve the linear equations
solution = np.linalg.solve(A, b)

# Plot the lines
x = np.linspace(-2, 6, 400)
y1 = (4 - 2*x)
y2 = (1 + x)

plt.plot(x, y1, label='2x + y = 4')
plt.plot(x, y2, label='x - y = 1')
plt.plot(solution[0], solution[1], 'ro', label=f'Solution: {solution}')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Solving Linear Equations')
plt.legend()
plt.show()
```

Singular Value Decomposition (SVD)

Singular Value Decomposition(SVD) is one of the great dimension-reduction algorithms in machine learning. It identifies the principal components and arranges them by rank. The top ranked components contribute greatly to the original array.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a matrix
A = np.array([[3, 1, 1], [-1, 3, 1]])

# Perform Singular Value Decomposition
U, S, VT = np.linalg.svd(A)

# Reconstruct the diagonal matrix of singular values
Sigma = np.zeros((A.shape[0], A.shape[1]))
Sigma[:len(S), :len(S)] = np.diag(S)

# Verify the decomposition
A_reconstructed = np.dot(U, np.dot(Sigma, VT))

# Print the results
print("Matrix A:\n", A)
print("U:\n", U)
print("Sigma:\n", Sigma)
print("V^T:\n", VT)
print("Reconstructed A:\n", A_reconstructed)

# Visualize the singular values
plt.plot(S, 'o-', label='Singular Values')
plt.xlabel('Index')
plt.ylabel('Singular Value')
plt.title('Singular Values of Matrix A')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc

# Load a sample image
image = misc.face(gray=True)  # Load a grayscale image

# Perform SVD
U, S, VT = np.linalg.svd(image, full_matrices=False)

# Reconstruct the image with reduced singular values
k = 50  # Number of singular values to keep
image_reconstructed = np.dot(U[:, :k], np.dot(np.diag(S[:k]), VT[:k, :]))

# Plot the original and reconstructed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f'Reconstructed Image with {k} singular values')
plt.imshow(image_reconstructed, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
